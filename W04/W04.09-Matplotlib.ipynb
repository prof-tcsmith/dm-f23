##### 1.1.1.1 Line Plot

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
plt.show()
```

##### 1.1.1.2 Scatter Plot

```python
plt.scatter(year, pop)
plt.show()
```

Example using numpy

```python
import numpy as np

np_height = np.round(np.random.normal(1.75, 0.2, 20), 2)
np_weight = np.round(np.random.normal(60.32, 15, 20), 2)

plt.scatter(np_height, np_weight)
plt.show()
```

### 1.3 Customization of Plots

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)

# here is the customizations
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0,2,4,6,8,10], ['0', '2B', '4B', '6B', '8B', '10B'])


plt.show()
```

```python
import pandas as pd

df = pd.read_csv('../../Data/world_population_data.csv', thousands=',') # you must specify the path to the data file. In this case, it is two folders up from the current folder inside the data folder
df
```

```python
df['billions'] = df.World_Population/ 1_000_000_000 
```

```python
import matplotlib.pyplot as plt
year = df['Year']
pop = df['billions']
plt.plot(year, pop)

# here is the customizations
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0,2,4,6,8,10], ['0', '2B', '4B', '6B', '8B', '10B'])


plt.show()
```
