### 2. Basic Plotting with Seaborn

Before diving into correlation matrices, let's briefly go over some of the basic plots you can create with Seaborn.

#### Scatter Plot

```python
# if running on colab, you may need to uncomment the following lines
# if you get an error saying that the module is not found
#!pip install matplotlib
#!pip install pandas
#!pip install ipympl
#!pip install seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

# Create scatter plot
sns.scatterplot(x=x, y=y)
plt.show()
```

#### Line Plot


```python
# Create line plot
sns.lineplot(x=x, y=y)
plt.show()
```

#### Histogram

```python
# Create histogram
sns.histplot(y, bins=5)
plt.show()
```

### 3. Correlation Matrix

#### What is a Correlation Matrix?

A correlation matrix is a table showing correlation coefficients between sets of variables. Each attribute of the dataset is compared with the other attributes to find out the correlation coefficient. This analysis allows you to see which pairs have the highest correlation, the pairs which are highly correlated represent the same variance to a certain degree.

#### Generating a Correlation Matrix with Seaborn

To generate a correlation matrix in Seaborn, you first need to calculate the correlation matrix using Pandas, and then you can use Seaborn's `heatmap` function to visualize it.

```python
import pandas as pd

# Sample Data
data = {'A': [1, 2, 3, 4, 5], 'B': [5, 4, 3, 2, 1], 'C': [1, 2, 1, 4, 3]}
df = pd.DataFrame(data)

# Calculate correlation matrix
correlation_matrix = df.corr()

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Create a paired plot matrix

# Sample data

data = {'A': [1, 2, 3, 4, 5], 'B': [5, 4, 3, 2, 1], 'C': [1, 2, 1, 4, 3]}

df = pd.DataFrame(data)

# Create pairplot
sns.pairplot(df)
plt.show()
```
